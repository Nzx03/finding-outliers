```python
statistics.variance(ages)
```

The difference is due to statistics  uses the sample variance formula (where there is divided by n-1) and np.var (population variance where there is divided by n)

```python
np.var(ages,axis=0)
```

```python
#to calculate manually
def variance(data,dof=0):
    n=len(ages)
    mean=sum(data)/n

    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-1)
    return variance
```

```python
variance(ages,dof=1)
```

```python
# to check normal distribution
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab) # for log normal (np.log(smaple))
    plt.show()



```

```python
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```
